# Poisson Distribution

describe the number of indepdent "success" or events in one interval or period. 

calculate the probabilities of various numbers of "successes" based on the mean number of successes. 

Let X be number of events or "success" in one interval,
    u = average number of events or "sucess" in one interval
    
=> the probability that x events occured in one interval follows Poisson distribution

p(X=x) = exp(-u)*(u^x)/(x!)

E(X) = u
var(X) = u

-----------------------
If X_i follows poisson(mu_i) distribution, then
sum(X_i) follows poisson(sum(mu_i)))

e.g., X1 ~ Poisson(mu1), X2 ~ Poisson(mu2), then X1+X2 ~ poisson(mu1+mu2)

-------------------------------- in Python ----------------------------
from scipy.stats import poisson

p(X=x) = pmf(x, mu, loc=0)

p(X<=x) = cdf(x, mu, loc=0)

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html

```python
from scipy.stats import poisson
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set()
```

```python
# plot the probability mass function of poisson
def plot_poisson(mu, pmf = True, cdf = False, **kwargs):
    fig, ax = plt.subplots(1, 1)

    # plot x with probability within [0.01, 0.99)
    x = np.arange(poisson.ppf(0.01, mu),
                    poisson.ppf(0.99, mu))   # poisson.ppf(1, mu) is infinity
    
    if pmf:
        ax.plot(x, poisson.pmf(x, mu), color = 'gray')
        ax.plot(x, poisson.pmf(x, mu), 'bo', ms=5, label='poisson pmf')
        ax.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=2, alpha=0.5)
    
    if cdf:
        ax.plot(x, poisson.cdf(x, mu), **kwargs)    
```

```python
mu = 10
plot_poisson(mu)
```

```python
mu = 10
plot_poisson(mu, pmf = True, cdf = True, color = 'orange', linestyle = ":")
```

```python
mu = 20
plot_poisson(mu)
```

Suppose you knew that the mean number of calls to a fire station on a weekday is 8. What is the probability that on a given weekday there would be 11 calls? This problem can be solved using the following formula based on the Poisson distribution:

```python
from scipy.stats import poisson
mu = 8
x = 11
prob_x = poisson.pmf(x, mu)
print(prob_x)
```

The mean number of defective products produced in a factory in one day is 21. What is the probability that in a given day there are exactly 12 defective products?

```python
mu = 21
x = 12
prob_x = poisson.pmf(x, mu)
print(prob_x)
```
