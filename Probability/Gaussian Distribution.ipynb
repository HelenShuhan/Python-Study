# Gaussian Distribution or Normal Distribution

```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from scipy.stats import norm  # standar normal distribution
```

```python
norm.pdf(2)
```

```python
norm.ppf(0.99)
```

```python
mu = 0
std = 1
norm.ppf(0.01, loc = mu, scale = std)
```

```python
def plot_normal(mu = 0, std = 1, pdf = True, cdf = False, *kwargs):
    # define the plot setting
    fig, ax = plt.subplots(1, 1)
    
    # define the region of x
    x = np.arange(norm.ppf(0.01, loc = mu, scale = std), 
                 norm.ppf(0.99, loc = mu, scale = std))
    
    # plot the normal distribution
    if pdf:
        ax.plot(x, norm.pdf(x, loc = mu, scale = std), color = 'gray')
        ax.plot(x, norm.pdf(x, loc = mu, scale = std), 'bo', ms =5)
   
    if cdf:
        ax.plot(x, norm.cdf(x, loc = mu, scale = std))
```

```python
plot_normal()
```

```python
mu = 22
std = 22/np.sqrt(80)
plot_normal(mu, std)
print(1- norm.cdf(20, loc = mu, scale = std))
```
